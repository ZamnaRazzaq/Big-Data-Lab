```python
from pyspark import SparkContext, SparkConf
```

```python
appName="lab1"
master="local[*]"
```

```python
conf = SparkConf().setAppName(appName).setMaster(master)
sc = SparkContext(conf=conf)
```

Task 1

```python
regularRDD = sc.parallelize([(1,2),(3,4),(3,6),(4,5)])
pairRDD= regularRDD.reduceByKey(lambda x,y : x + y)
pairRDD.collect()
for k, v in pairRDD.collectAsMap().items():
  print("key", k, "has", v, "counts")
```

```python
baseRDD = sc.textFile("shakespeare.txt")
```

```python
rdd1 = baseRDD.flatMap(lambda line: line.split(" "))
for word in rdd1.collect():
    print(word)
```

```python
RDD = baseRDD.flatMap(lambda word: word.split(" "))
rdd4=RDD.count()
print("Words in file: ",rdd4)
```

```python
splitRDD = baseRDD.flatMap(lambda x: x.split())
rdd2=splitRDD.map(lambda word: (word, 1))
rdd3=rdd2.reduceByKey(lambda x, y: x + y)
rdd3.collect()
```

```python
rdd5= RDD.map(lambda word: word.lower())
```

```python
stopwords = ['a', 'an', 'the', 'and', 'but', 'or', 'for', 'nor', 'on', 'at', 'to', 'from', 'by', 'so', 'yet']
rdd6 = rdd5.filter(lambda word: word not in stopwords)
```

```python
rdd7=rdd6.map(lambda word: (word, 1))
rdd8=rdd7.reduceByKey(lambda x, y: x + y)
rdd8.collect()
```

```python
rdd9=rdd8.map(lambda word:(word[1] , word[0]))
```

```python
rdd9.sortByKey(ascending=False).take(10)
```
